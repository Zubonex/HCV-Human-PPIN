## Indentification of Pure Hub by means of Degreee of Network
Pure hub proteins in the HCV–host protein–protein interaction network were identified using degree centrality, defined mathematically as:
$$
C_D(v) = \deg(v)
$$
where $C_D(v)$ represents the degree centrality of a protein $v$, and $deg(v)$ is the total number of direct interactions (edges)
connected to that protein.
Proteins with degree values satisfying:
$$
\deg(v) \gg \overline{\deg}
$$
(where $ \overline{\deg}$ is the average degree of the network) were classified as <b>pure hubs</b>).
In the case of HCV, viral proteins such as Core, E1, E2, NS3, NS5A, and NS5B showed significantly high degree values, indicating extensive interactions with host proteins. These high-degree proteins act as central hubs that are essential for viral entry, replication, and immune evasion.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Analysis.csv')


protein_column = 'display name'
centrality_column = 'Degree'


df_sorted = df.sort_values(by=centrality_column, ascending=False)

df_top_10 = df_sorted.head(10)


plt.figure(figsize=(10, 8))

plt.barh(
    df_top_10[protein_column][::-1],    
    df_top_10[centrality_column][::-1], 
    color='darkblue'
)


plt.title('Top 10 Protein Hubs by Degree Centrality', fontsize=16)
plt.xlabel('Degree Centrality Value')
plt.ylabel('Protein Name')
plt.gca().tick_params(axis='y', length=0) 
plt.tight_layout()


plt.savefig('top_10_horizontal_bar_chart.png')
print("Plot saved as top_10_horizontal_bar_chart.png")
    
```

### Interpretation of Top 10 Protein Hubs by Degree Centrality
1. <b>NS5A</b> shows the highest degree centrality, indicating it is the most interconnected and dominant hub in the HCV–host PPIN.

2. <b>ARFP/F, NS4A</b>, and **NS4B** also have high degree values, highlighting their major roles in viral replication and modulation of host machinery.

3. **Core p19** appears as a key structural hub interacting with multiple host proteins.

4. **RNA-directed RNA polymerase (NS5B)** shows strong connectivity, reflecting its essential function in viral RNA synthesis.

5. **Serine protease NS3** is an important hub due to its involvement in polyprotein processing and immune evasion.

6. **Envelope glycoprotein E2** displays moderate hub characteristics, consistent with its role in viral entry.

7. Host proteins **TBXT** and **CDSN** show lower but notable degree values, suggesting auxiliary roles in the interaction network.

Overall, the graph shows that HCV non-structural proteins dominate the top hubs, highlighting their central role in regulating the virus–host interaction network.

## Indefication of Spreader Nodes by means of Betweenness Centrality measure
Spreader nodes in the HCV–host protein interaction network were identified using the betweenness centrality measure, which quantifies how frequently a protein lies on the shortest paths between other proteins. Proteins with high betweenness values act as bridges that connect different clusters of the network and facilitate the flow of biological information. Unlike hub proteins, which have many direct interactions, spreader nodes may have fewer connections but play a critical structural role in maintaining communication between otherwise disconnected regions of the network. Mathematically, betweenness centrality is expressed as:
$$
C_B(v) = \sum_{s \neq v \neq t} \frac{\sigma_{st}(v)}{\sigma_{st}}
$$

where $\sigma_{st}$ represents the total number of shortest paths between nodes $s$ and $t$, and $\sigma_{st}(v)$ is the number of those paths that pass through protein $v$. Proteins scoring high on this measure were classified as **spreader proteins**, indicating their importance in regulating viral–host interactions and coordinating cross-pathway communication.

```python

df = pd.read_csv('Analysis.csv')


protein_column = 'display name'
centrality_column = 'BetweennessCentrality'


df_sorted = df.sort_values(by=centrality_column, ascending=False)

df_top_10 = df_sorted.head(10)


plt.figure(figsize=(10, 8))

plt.barh(
    df_top_10[protein_column][::-1],    
    df_top_10[centrality_column][::-1],
    color='darkblue'
)


plt.title('Top 10 Protein Hubs by Betweenness Centrality', fontsize=16)
plt.xlabel('Betweenness Centrality Value')
plt.ylabel('Protein Name')
plt.gca().tick_params(axis='y', length=0) 
plt.tight_layout()


plt.savefig('Betweenness_top_10.png')
print("Plot saved as top_10_horizontal_bar_chart.png")
    
```

## Interpretation of Top 10 Spreader Proteins by Betweenness Centrality
The bar chart shows the top spreader proteins identified using betweenness centrality, which highlights proteins that act as bridges within the HCV–host interaction network. NS5A has the highest betweenness value, making it the main connector in the network, followed by **ARFP/F** and **NS3**. Other proteins such as **Core** **p19**, **E2**, **NS4A**, and **NS4B** also show moderate bridging roles, while **FUT1** and **TBXT** appear as minor spreader nodes. Overall, the plot indicates that HCV non-structural proteins dominate the key spreader positions.
