```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Analysis.csv')


protein_column = 'display name'
centrality_column = 'Degree'


df_sorted = df.sort_values(by=centrality_column, ascending=False)

df_top_10 = df_sorted.head(10)


plt.figure(figsize=(10, 8))

plt.barh(
    df_top_10[protein_column][::-1],    
    df_top_10[centrality_column][::-1], 
    color='darkblue'
)


plt.title('Top 10 Protein Hubs by Degree Centrality', fontsize=16)
plt.xlabel('Degree Centrality Value')
plt.ylabel('Protein Name')
plt.gca().tick_params(axis='y', length=0) 
plt.tight_layout()


plt.savefig('top_10_horizontal_bar_chart.png')
print("Plot saved as top_10_horizontal_bar_chart.png")
    
```
